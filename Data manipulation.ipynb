## Merging Dataframes

```python
#Using concat() to merge dataframes

df1=pd.DataFrame({"A": ["A1", "A2", "A3", "A4"], "B":["B1", "B2", "B3", "B4"], "C":["C1", "C2", "C3", "C4"]})
df2=pd.DataFrame({"D":["D1","D2","D3","D4"], "E":["E1","E2", "E3","E4"], "F":["F1", "F2", "F3","F4"]})
df3= pd.DataFrame({"G":["G1","G2","G3","G4"],"H":["H1","H2","H3","H4"]})
```

```python
#Using concat to merge rows of df
Frame=pd.concat([df1,df2,df3])
Frame
```

```python
#Using keys to recognise the concatenated dfs
Frame=pd.concat([df1,df2,df3], keys=["X","Y","Z"])
```

```python
Frame
```

```python
#To get the Second dataframe

Frame.loc["Y"]
```

Using join=outer/inner

```python
#Using join inner

Frame=pd.concat([df1,df3], axis=1, join="inner")
```

```python
Frame
```

Using Merge

```python
df_a=pd.DataFrame({"Subject_id":["1","2","3","4","5"], "First_name":["Edward", "Peter", "Alice", "Frank", "Queen"], "Second_name":["Ade", "Dan", "Joyce", "Tony", "Collins"]})

df_b=pd.DataFrame({"Subject_id":["4", "5", "6", "7", "8"], "First_name":["Grey", "West", "Goerge", "Bran", "Ken"], "Second_name":["kate", "Black", "Liz", "Angel", "Faith"]})
```

```python
#Using merge to merge the dataframe with similar subject_id

pd.merge(df_a, df_b, on="Subject_id")
```

```python
#Merging using Outer Joining

pd.merge(df_a, df_b, on="Subject_id", how="outer")
```

```python
#Merging using inner 

pd.merge(df_a, df_b, on="Subject_id", how="inner")
```

```python
#Merging using right 

pd.merge(df_a, df_b, on="Subject_id", how="right")
```

```python
#Merging using left

pd.merge(df_a, df_b, on="Subject_id", how="left")
```

Using apply function to implementcondition individually on each row/column

```python
data["Item_MRP"]
```

```python
#Declaring a function to make the item price greater than 200 to be equal to 200
def clip_price(price):
    if price>200:
        price=200
    return price

data["Item_MRP"].apply(lambda X: clip_price(X))
```

```python
#To label encode using apply function
data["Outlet_Location_Type"]
```

```python
#Declaring a function to label encode the tiers in Outlet_Location_Type column

def label_encode(city):
    if city=="Tier 1":
        label= 0
    elif city=="Tier 2":
        label= 1
    else:
        label=2
    return label
data["Outlet_Location_Type"].apply(lambda X: label_encode(X))
```

```python
#Declaring a function to label encode outlet type column
def label_encode(Outlet):
    if Outlet=="Supermarket Type1":
        label= 0
    elif Outlet=="Supermarket Type2":
        label= 1
    else:
        label=2
    return label
data["Outlet_Type"].apply(lambda X: label_encode(X))
```

```python
#Checking the max of the outlet sales

data["Item_Outlet_Sales"].max()
```

```python
#Checking minimum of the outlet sales 
data["Item_Outlet_Sales"].min()
```

## Aggregating Data

```python
#Resetting index after dropping

data=data.reset_index(drop=True)
```

```python
data
```

### a. What is the mean price of each item?
First step is to group the item by Item type column

```python
price_by_Item=data.groupby("Item_Type")
price_by_Item.first()
```

Then look for the mean of the types

```python
price_by_Item.Item_MRP.mean()
```

```python
#Groupping by more than 1 column
price_by_Item1=data.groupby(["Item_Type", "Item_Fat_Content"])
price_by_Item1.first()
```

### b. How are the Outlet size distributed based on the city type?
Using Crosstab

```python
#Create a dataframe to check how outlet size affected outlet type using crosstab

pd.crosstab(data["Outlet_Size"], data["Outlet_Type"], margins=True)
```

### c. How are the sales changing per year?
Using piovttable

```python
#Create a pivot table to check how the price changes with year
Sales_per_year=pd.pivot_table(data, index="Outlet_Establishment_Year", values="Item_Outlet_Sales")
Sales_per_year
```

```python
#Visualising the result
import matplotlib.pyplot as plt
Sales_per_year.plot()
```

```python
#Passing several columns to pivot table
pd.pivot_table(data, index=["Outlet_Establishment_Year", "Outlet_Location_Type", "Outlet_Size"], values="Item_Outlet_Sales")
```

```python
#Performing different aggregation using aggfunc
pd.pivot_table(data, index=["Outlet_Establishment_Year", "Outlet_Location_Type", "Outlet_Size"], values="Item_Outlet_Sales", aggfunc=[np.mean, np.median, min, max, np.std,])
```
